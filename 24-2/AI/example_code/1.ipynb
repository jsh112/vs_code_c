```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(seed=1)
X_min = 4
X_max = 30
X_n = 16
X = 5 + 25 * np.random.rand(X_n)
```

```python
Prm_c = [170, 108, 0.2]
T = Prm_c[0] - Prm_c[1] * np.exp(-Prm_c[2] * X) + 4 * np.random.randn(X_n)
```

```python
print(np.round(X, 2))
print(np.round(T, 2))
```

```python
import numpy as np

# Define matrix A and vector b
A = np.array([[np.sum(X**2), np.sum(X)], 
              [np.sum(X), len(X)]])
b = np.array([np.sum(X*T), np.sum(T)])

# Solve the linear system A * x = b
x = np.linalg.solve(A, b)
```

```python
for i in range(len(X)):
    plt.scatter(np.round(X[i],2), np.round(T[i],2),color='blue')
X_line = np.linspace(min(X), max(X), 100)  
T_line = x[0] * X_line + x[1]
plt.plot(X_line, T_line, color='red')  
plt.grid()
plt.show()
```

```python
#  
theta_0 = 0.0  
theta_1 = 0.0  
alpha = 0.002
n_iterations = 10000  
error_threshold = 0.01
n = len(T)  

for iteration in range(n_iterations):
    y_pred = theta_0 + theta_1 * X  
    error = y_pred - T 
    cost = (1/n) * np.sum(error**2)  
    
    #  
    gradient_0 = (2/n) * np.sum(error)  
    gradient_1 = (2/n) * np.sum(error * X)  
    
    #  
    theta_0 -= alpha * gradient_0
    theta_1 -= alpha * gradient_1
    
    if cost < error_threshold:
        print(f"Converged after {iteration + 1} iterations.")
        break

#  
print(f'(theta_0): {theta_0:.2f}')
print(f'(theta_1): {theta_1:.2f}')

#     
plt.scatter(X, T, color='blue', label='data point')
plt.plot(X, theta_0 + theta_1 * X, color='red', label='Linear Regression')
plt.xlabel('x')
plt.ylabel('y')
plt.title('L')
plt.grid()
plt.legend()
plt.show()
```
