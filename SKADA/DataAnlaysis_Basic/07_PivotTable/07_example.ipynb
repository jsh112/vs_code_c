```python
from IPython.display import Image
import numpy as np
import pandas as pd
import seaborn as sns
import warnings

# 경고 무시
warnings.filterwarnings('ignore')

df = sns.load_dataset('tips')
df
```

#### 그룹별 통계량 확인

- DataFrame에 .groupby(column) + 통계함수(max, min, mean, median...)로 그룹별 통계량을 확인할 수 있습니다.
- 통계 결과는 통계계산이 가능한 수치형(numerical) column에 대해서만 산출합니다.

```python
# 평균
df.groupby('sex').mean(numeric_only=True)
```

```python
# 분산
df.groupby('sex').var(numeric_only=True)
```

`.agg()`에 적용할 수 있는 통계함수 문자열 표

```python
data = {
    '함수' : ['count', 'sum', 'mean', 'median', 'var, std', 'min, max', 'unique, nunique', 'prod', 'first, last'],
    '내용' : ['데이터의 개수', '합', '평균', '중앙값', '분산, 표준편차', '최대, 최소', '고윳값, 고윳값의 개수', '곱', '첫째, 마지막 값']
}
pd.DataFrame(data)
```

`.agg()`를 활용하여 다중 통계량을 구할 수 있습니다.

```python
# 다중 통계 적용
df.groupby('sex').agg(['mean', 'var'])
```
